```python
#import lg dataset (two group columns in an excel file)
df = pd.read_excel('ta data.xlsx', header=0)
#check the first few rows of the dataframe
print(df.head())
```

```python
two_groups_unpaired = dabest.load(df, idx=("WT", "RC-DTX"), resamples=5000, ci = 95)
```
